Neste projeto o estudante deverá individualmente construir um script em Python que receba do usuário duas variáveis de valores referentes a variáveis quantitativas e devolva como resposta a seguintes medidas de estatística descritiva para cada variável informada:

* Média aritmética;
* Mediana;
* Moda (se houver);
* Variância amostral;
* Desvio Padrão amostral;
* Amplitude total;
* Q1 (primeiro quartil);
* Q3 (terceiro quartil);
* AIQ (amplitude interquartílica);
* Limite inferior (LI do boxplot);
* Limite superior (LS do boxplot);
* Identificação e classificação de possíveis outliers na amostra;
* Coeficiente de Correlação de Pearson ( r );
* Coeficiente de determinação ( r ao quadrado);
* Coeficiente a ou alpha da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário;
* Coeficiente b ou beta da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário;


# **Requisitos obrigatórios:**

* O Programa deve conter no mínimo as seguintes estruturas:

```python
# Solicitar lista de números ao usuario separados por espaço.
while True:
    try:
        lista1 = input('Informe uma lista de números separados por espaço: ').split()
        lista2 = input('Informe outra lista de números separados por espaço com a mesma quantidade de números da anterior: ').split()
        if len(lista1) == len(lista2):
            break
        else:
            print('As listas não possuem o mesmo tamanho. Tente novamente.')
    except:
        pass

# Transformar os dados da listas de str para float.

for i in range(len(lista1)):
    lista1[i] = float(lista1[i])

for i in range(len(lista2)):
    lista2[i] = float(lista2[i])

# Calculo da Média.

media1 = sum(lista1) / len(lista1)
media2 = sum(lista2) / len(lista2)

# Calculo da Mediana.

lista1 = sorted(lista1)
lista2 = sorted(lista2)

qtd1 = len(lista1)
qtd2 = len(lista2)

if qtd1 % 2 == 0:
    mediana1 = (lista1[qtd1 // 2 - 1]) + (lista1[qtd1 // 2]) / 2
else:
    mediana1 = lista1[(len(lista1) + 1) // 2 - 1]


if qtd2 % 2 == 0:
    mediana2 = (lista2[qtd2 // 2 - 1]) + (lista2[qtd2 // 2]) / 2
else:
    mediana2 = lista2[(len(lista2) + 1) // 2 - 1]


# Achando a Moda (Script retirado da internet)

contagem1 = {}
max_contagem1 = 0
moda1 = []

for valor1 in lista1:
    if valor1 in contagem1:
        contagem1[valor1] += 1
    else:
        contagem1[valor1] = 1

    if contagem1[valor1] > max_contagem1:
        max_contagem1 = contagem1[valor1]

for valor1 in contagem1:
    if contagem1[valor1] == max_contagem1:
        moda1.append(valor1)


contagem2 = {}
max_contagem2 = 0
moda2 = []

for valor2 in lista2:
    if valor2 in contagem2:
        contagem2[valor2] += 1
    else:
        contagem2[valor2] = 1

    if contagem2[valor2] > max_contagem2:
        max_contagem2 = contagem2[valor2]

for valor2 in contagem2:
    if contagem2[valor2] == max_contagem2:
        moda2.append(valor2)


# Calculando a Variancia.

quad_v1 = []
for i in range(len(lista1)):
    quad_v1.append((lista1[i] - media1)**2)
variancia1 = sum(quad_v1) / (len(lista1)-1)

quad_v2 = []
for i in range(len(lista2)):
    quad_v2.append((lista2[i] - media2)**2)
variancia2 = sum(quad_v2) / (len(lista2)-1)


# Calculando o Desvio Padrão.

desv_pad1 = variancia1 ** 0.5

desv_pad2 = variancia2 ** 0.5


# Achando a Amplitude Total.

ampli_total1 = max(lista1) - min(lista1)

ampli_total2 = max(lista2) - min(lista2)


# Achando o Quartil 1.

ind_q1_l1 = int(qtd1 * 0.25)
q1_l1 = lista1[ind_q1_l1]


ind_q1_l2 = int(qtd2 * 0.25)
q1_l2 = lista2[ind_q1_l2]


# Achando o Quartil 3.

ind_q3_l1 = int(qtd1 * 0.75)
q3_l1 = lista1[ind_q3_l1]


ind_q3_l2 = int(qtd2 * 0.75)
q3_l2 = lista2[ind_q3_l2]


# Achando a Amplitude Interquartilica.

aiq_l1 = q3_l1 - q1_l1

aiq_l2 = q3_l2 - q1_l2


# Achando o Limite Inferior.

li_l1 = q1_l1 - (1.5 * aiq_l1)

li_l2 = q1_l2 - (1.5 * aiq_l2)


# Achando o Limite Superior.

ls_l1 = q3_l1 + (1.5 * aiq_l1)

ls_l2 = q3_l2 + (1.5 * aiq_l2)


# Verificando o Coeficiente de Correlação de Pearson ( r ).

soma_l1 = sum(lista1)
soma_l2 = sum(lista2)
l1_quad = 0
l2_quad = 0
l1_l2 = 0

for i in range(len(lista1)):
    l1_quad = l1_quad + (lista1[i]**2)
    l2_quad = l2_quad + (lista2[i]**2)
    l1_l2 = l1_l2 + (lista1[i]*lista1[i])


numerador = l1_l2 - (((soma_l1 * soma_l2)) / len(lista1))
denominador = ((l1_quad - ((soma_l1**2) / len(lista1))) * (l2_quad - ((soma_l2**2) / len(lista1)))) ** 0.5
r = numerador / denominador


# Achando o Coeficiente de determinação ( r ao quadrado).

r_quad = r**2


# Achando o Coeficiente b ou beta da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário.

beta = l1_l2 - (((soma_l1 * soma_l2)) / len(lista1)) / (l1_quad - ((soma_l1**2) / len(lista1)))


# Achando o Coeficiente a ou alpha da equação de regressão linear simples para as duas variáveis escolhidas pelo usuário.

alpha = media2 - media1 * beta

# Prints dos resultados
print('                                         ')
print('------------DADOS DA LISTA 1------------')
print(f'MÉDIA: {media1:.2f}')
print(f'MEDIANA: {mediana1:.2f}')
if len(moda1) == 1:
    print(f'MODA: {moda1[0]}')
else:
    print('NÃO POSSUI MODA')
print(f'VARIANCIA: {variancia1:.2f}')
print(f'DESVIO PADRÃO: {desv_pad1:.2f}')
print(f'AMPLITUDE TOTAL: {ampli_total1}')
print(f'QUARTIL 1: {q1_l1}')
print(f'QUARTIL 3: {q3_l1}')
print(f'AMPLITUDE INTERQUARTILICA: {aiq_l1}')
print(f'LIMITE INFERIOR: {li_l1}')
print(f'LIMITE SUPERIOR: {ls_l1}')
print('-----------------------------------------')
print('                                         ')
print('------------DADOS DA LISTA 2------------')
print(f'MÉDIA: {media2:.2f}')
print(f'MEDIANA: {mediana2:.2f}')
if len(moda2) == 1:
    print(f'MODA: {moda2[0]}')
else:
    print('NÃO POSSUI MODA')
print(f'VARIANCIA: {variancia2:.2f}')
print(f'DESVIO PADRÃO: {desv_pad2:.2f}')
print(f'AMPLITUDE TOTAL: {ampli_total2}')
print(f'QUARTIL 1: {q1_l2}')
print(f'QUARTIL 3: {q3_l2}')
print(f'AMPLITUDE INTERQUARTILICA: {aiq_l2}')
print(f'LIMITE INFERIOR: {li_l2}')
print(f'LIMITE SUPERIOR: {ls_l2}')
print('-----------------------------------------')

print('                                                    ')

print('------------------------MEDIDAS DE DISPERSÃO------------------------')
print(f'COEFICIENTE DA CORRELAÇÃO DE PEARSON: {r:.2f}')
print(f'COEFICIENTE BETA DA EQUAÇÃO DE REGRESSÃO LINEAR SIMPLES: {beta:.2f}')
print(f'COEFICIENTE ALPHA DA EQUAÇÃO DE REGRESSÃO LINEAR SIMPLES: {alpha:.2f}')

# Verificação de possiveis outliers

for i in range(len(lista1)):
    if lista1[i] < li_l1:
        print(f'O VALOR {lista1[i]:.2f} DA LISTA 1 É UM OUTLIER. O LIMITE INFERIOR É {li_l1:.2f}')
    elif lista1[i] > ls_l1:
        print(f'O VALOR {lista1[i]:.2f} DA LISTA 1 É UM OUTLIER. O LIMITE SUPERIOR É {ls_l1:.2f}')
    else:
        pass

for i in range(len(lista2)):
    if lista2[i] < li_l2:
        print(f'O VALOR {lista2[i]:.2f} DA LISTA 2 É UM OUTLIER. O LIMITE INFERIOR É {li_l2:.2f}')
    elif lista2[i] > ls_l2:
        print(f'O VALOR {lista2[i]:.2f} DA LISTA 2 É UM OUTLIER. O LIMITE SUPERIOR É {ls_l2:.2f}')
    else:
        pass

print('--------------------------------------------------------------------')
```
